### 문제 3의 발전된 예시

- Scatter plot 한 부분을 더 더양한 메소드와 함수 등을 이용해서 시각화에 있어서 더 유리하게 변경할 수 있습니다.
- 예를 들어 과밀된 곳은 색깔을 더 빨갛게 한다던지, 실제 지도를 배경으로 두어 Plot을 한다던지 하면 dataset을 더 이해하기 쉽게 시각화할 수 있습니다.

```
from pathlib import Path
import urllib.request

# 1. 데이터 불러오기
housing = pd.read_csv('sample_data/california_housing_train.csv')

# 2. 캘리포니아 지도 이미지 다운로드 설정
IMAGES_PATH = Path() / "images"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)
filename = "california.png"

# 깃허브에서 이미지 다운로드
if not (IMAGES_PATH / filename).is_file():
    homl3_root = "https://github.com/ageron/handson-ml3/raw/main/"
    url = homl3_root + "images/end_to_end_project/" + filename
    print("Downloading", filename)
    urllib.request.urlretrieve(url, IMAGES_PATH / filename)

# 3. 위도/경도를 이용한 구역별 인구 산포도 그리기
housing.plot(kind="scatter",
             x="longitude",
             y="latitude",
             s=housing["population"] / 100, label="Population",
             c="median_house_value",
             cmap="turbo",
             colorbar=True,
             figsize=(10, 7),
             alpha=0.4)

# 4. 다운로드된 캘리포니아 지도 겹쳐서 표시
california_img = plt.imread(IMAGES_PATH / filename)
axis = [-124.55, -113.8, 32.45, 42.3] # x축, y축 눈금 (지도의 실제 위도/경도 위치)
plt.axis(axis)
plt.imshow(california_img, extent=axis)

plt.show()
```
